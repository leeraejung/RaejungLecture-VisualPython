# 1. Import python package

```python
# Visual Python: Data Analysis > Import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# 2. Load files

```python
# Visual Python: Data Analysis > File
df_iris = pd.read_csv('data/iris.csv')
df_iris
```

# 3. Subset

## 3.1 subset

```python
# Visual Python: Data Analysis > Subset
df_iris[(df_iris['sepal_length'] < 4.5)][['sepal_length','sepal_width','variety']]
```

## 3.2 loc (location)

```python
# Visual Python: Data Analysis > Subset
df_iris.loc[(df_iris['variety'] == 'Virginica')&(df_iris['sepal_length'] < 6), ['sepal_length','sepal_width','variety']]
```

## 3.3 iloc (Integer location)

```python
# Visual Python: Data Analysis > Subset
df_iris.iloc[0:10, [1,2,4]]
```

## 3.4 query

```python
# Visual Python: Data Analysis > Subset
df_iris.query("(`sepal_length` < 6)&(`variety` == 'Virginica')")[['sepal_length','sepal_width','variety']]
```
